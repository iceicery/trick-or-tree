```python
import geopandas as gpd
import pandas as pd
import numpy as np
import requests
import os
import time
import wikipedia as wp
```

```python
gdf = gpd.read_file('NEWal_champTree_leafimg_wiki.geojson')
noimg = []
for index, row in gdf.iterrows():
    if row['image_path'] is None:
        noimg.append(index)
noimg = np.array(noimg)
```

```python
imagepath = '/home/ctchen/work_scripts/sideprojects/HATCH2021/trick-or-tree/dataset/images/wiki/'
for i in noimg:
    print(gdf.loc[i,'scientific_name'])
    tree = wp.page(gdf.loc[i,'scientific_name'])
    time.sleep(1)
    imgurl = tree.images[0]
    img_data = requests.get(imgurl).content
    sciname = gdf.scientific_name[i].lower().replace(' ','_')
    os.system('mkdir ' + imagepath + sciname)
    image_name = imagepath + sciname + '/' + sciname + '.' + imgurl.split('.')[-1]
    with open(image_name, 'wb') as handler:
        handler.write(img_data)
        print('downloaded image for ' + sciname + ' from wiki and saved as \n' + image_name)
        gdf.loc[i, 'image_path'] = image_name
```
