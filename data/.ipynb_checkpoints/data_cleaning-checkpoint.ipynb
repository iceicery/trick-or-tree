```python
import geopandas as gpd
import pandas as pd
import numpy as np
```

```python
import re
def delnums(my_str):
    return re.sub('[0-9]+', '', my_str)
```

```python
import tabula
```

```python
file = "Champion_Trees.pdf"

table = tabula.read_pdf(file,pages=3)
df = table[0]
names = df.loc[1::3]['Unnamed: 0'].values
primary_name = []
secondary_name = []
for i, s in enumerate(names):
    newstr = s.split(',')
    primary_name.append(newstr[0])
    if len(newstr) == 2:
        secondary_name.append(newstr[1])
    else:
        secondary_name.append('N/A')
```

```python
arr = df.loc[2::3]['Unnamed: 0']
scientific_name = []
Year_Named = []
for i, s in enumerate(arr):
    try:
        Year_Named.append(int(s[-4:]))
    except:
        Year_Named.append(np.nan)
    newstr = delnums(s)
    scientific_name.append(newstr.strip())
```

```python
circumference = df.loc[2::3]['(inches)'].values
height = df.loc[2::3]['(feet)'].values
spread = df.loc[2::3]['(feet).1'].values
total_points = df.loc[2::3]['Points'].values
county =  df.loc[2::3]['Unnamed: 1'].values
Year_remeasured = df.loc[2::3]['Year'].astype(int).values
```

```python
arr = df.loc[2::3]['Nominator (N)'].values
owner = []
nominator = []
for i, s in enumerate(arr):
    s = s.replace('–',',')
    s = s.replace('-',',')
    #if '-O' in s: s=s.replace('-O',' - O')
    newlist = s.split(',')
    owner.append(newlist[0].strip())
    if s.strip()[-1] == 'N':
        nominator.append(newlist[0].strip())
    else:
        nominator.append('TBD')
```

```python
arr = df.loc[3::3]['Nominator (N)'].reset_index().drop('index',axis=1)
for i in np.where(arr.notnull())[0]:
    s = arr['Nominator (N)'][i]
    s = s.replace('–',',')
    s = s.replace('-',',')
    newstr = s.split(',')
    nominator[i] = newstr[0].strip()
```

```python
mdf = pd.DataFrame(
{
    'species':primary_name,
    'subspecies':secondary_name,
    'scientific_name':scientific_name,
    'year_named':Year_Named,
    'circumference_in':circumference,
    'height_ft':height,
    'spread_ft':spread,
    'total_points':total_points,
    'county':county,
    'year_remeasured':Year_remeasured,
    'owner':owner,
    'nominator':nominator
})
```

```python
file = "Champion_Trees.pdf"


for p in range(4,11):
    print('Page ' + str(p))
    table = tabula.read_pdf(file,pages=p)
    df = table[0]

    names = df.loc[1::3]['Unnamed: 0'].values
    primary_name = []
    secondary_name = []
    for i, s in enumerate(names):
        newstr = s.split(',')
        primary_name.append(newstr[0])
        if len(newstr) == 2:
            secondary_name.append(newstr[1])
        else:
            secondary_name.append('N/A')
    arr = df.loc[2::3]['Unnamed: 0']
    scientific_name = []
    Year_Named = []
    for i, s in enumerate(arr):
        try:
            Year_Named.append(int(s[-4:]))
        except:
            Year_Named.append(np.nan)
        newstr = delnums(s)
        scientific_name.append(newstr.strip())

        
        
        
    circumference = df.loc[2::3]['(inches)'].values
    height = df.loc[2::3]['(feet)'].values
    spread = df.loc[2::3]['(feet).1'].values
    total_points = df.loc[2::3]['Points'].values
    county =  df.loc[2::3]['Unnamed: 1'].values
    Year_remeasured = df.loc[2::3]['Year'].astype(int).values

    arr = df.loc[2::3]['Nominator (N)'].values
    owner = []
    nominator = []
    for i, s in enumerate(arr):
        s = s.replace('–',',')
        s = s.replace('-',',')
        #if '-O' in s: s=s.replace('-O',' - O')
        newlist = s.split(',')
        owner.append(newlist[0].strip())
        if s.strip()[-1] == 'N':
            nominator.append(newlist[0].strip())
        else:
            nominator.append('TBD')
    arr = df.loc[3::3]['Nominator (N)'].reset_index().drop('index',axis=1)        
    for i in np.where(arr.notnull())[0]:
        s = arr['Nominator (N)'][i]
        s = s.replace('–','-')
        newstr = s.split(' - ')
        nominator[i] = newstr[0]
    newdf = pd.DataFrame(
    {
        'species':primary_name,
        'subspecies':secondary_name,
        'scientific_name':scientific_name,
        'year_named':Year_Named,
        'circumference_in':circumference,
        'height_ft':height,
        'spread_ft':spread,
        'total_points':total_points,
        'county':county,
        'year_remeasured':Year_remeasured,
        'owner':owner,
        'nominator':nominator
    })    
    mdf = pd.concat((newdf,mdf))
```

```python
mdf = mdf.reset_index().drop('index',axis=1)
mdf['scientific_name'] = [i.replace(',','') for i in mdf.scientific_name.values]
mdf['sci_lc'] = [i.lower().strip() for i in mdf.scientific_name.values]
len(np.unique(np.array(mdf.sci_lc.values)))
```

```python
# check leafsnap image for AL champion tree species.
leafimg = pd.read_csv('leafsnap-dataset-images.txt',sep='\t')
leafimg = leafimg.drop_duplicates('species').reset_index().drop('index',axis=1)
leafimg['sci_lc'] = [i.lower().strip() for i in leafimg.species.values]
```

```python
merged = pd.merge(mdf, leafimg, on="sci_lc",how='left')
merged
```

```python
merged.to_csv('alchamptree_with_leafimg.csv')
```

```python
species = merged.scientific_name.values
```

```python
import wikipedia

possible_entries = wikipedia.search(species[0])
```

```python
wikipedia.summary(species[0])
```

```python
# there was a typo in scientific names.
np.where(species == 'Pinus echinate')
```

```python
# identified typos in these ... updating
species[39] = 'Pinus echinata'
species[110] = 'Chamaecyparis thyoides'
species[129] = 'Crataegus spathulata'
species[131] = ' Cotinus obovatus'
species[148] = 'Sideroxylon lycioides' # updated to the synonym of Bumelia lycioides.
```

```python

summaries = []
badid = [91,137]
for i, s in enumerate(species):
    if i not in badid:
        try:
            summaries.append(wikipedia.summary(s,auto_suggest=False))
            print(i, s)
        except wikipedia.exceptions.DisambiguationError as e:
            print(i, e.options)
            summaries.append(wikipedia.summary(e.options[0],auto_suggest=False))            
    elif i == 91:
        summaries.append('wikipedia page not found.')
        print(i, 'no wiki')
    elif i == 137:
        summaries.append('Not a valid scientific name.')
        print(i, 'not a sci name')        
```

```python
merged['wiki_summary'] = summaries
```

```python
for index, row in merged.iterrows():
    c = row['county'].strip()
    merged.loc[index, 'Longitude'] = float(df[df.county == c].lon.values[0])
    merged.loc[index, 'Latitude'] = float(df[df.county == c].lat.values[0])
    
```

```python
gdf = gpd.GeoDataFrame(
    merged, geometry=gpd.points_from_xy(merged.Longitude, merged.Latitude))
```
